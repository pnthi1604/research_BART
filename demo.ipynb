```python
import matplotlib.pyplot as plt
```

```python
# figures
def draw_graph(title, xlabel, ylabel, all_data):
    print("Drawing graph...")
    print(all_data)
    for data, info in all_data:
        x = list(range(len(data)))
        plt.plot(x, data, label=info)
        # add multiple legends
        plt.legend()

    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
    plt.close()
```

```python
draw_graph(
    title='Test Graph',
    xlabel='x-axis',
    ylabel='y-axis',
    all_data=[(list(range(10)), 'duong 1'), (list(range(10, 0, -1)), 'duong 2')]
)
```
