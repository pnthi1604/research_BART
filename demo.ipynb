```python
import matplotlib.pyplot as plt
import pandas as pd
import torch
import json
```

```python
def draw_multi_graph(title, xlabel, ylabel, all_data, steps):
    try:
        for data, info in all_data:
            plt.plot(steps, data, label=info)
            # add multiple legends
            plt.legend()

        plt.title(title)
        plt.xlabel(xlabel)
        plt.ylabel(ylabel)
        plt.yscale('log')
        plt.grid(True)
        plt.show()
        plt.close()
    except Exception as e:
        print(e)
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 8, 27, 64, 125]

x1 = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
```

```python
draw_multi_graph(
    title='Test',
    xlabel='X',
    ylabel='Y',
    all_data=[(y, 'y'), (y1, 'y1')],
    steps=x
)
```
